## Statistical Functions

Let's look into mean

$$\bar{x} = \frac{\sum{x}}{n}$$

```python
mean_age = sum(age) / len(age)
print(mean_age)
```

```python

plt.title("Age of Fifa Players")
plt.xlabel("Player Number" + u"\u2192")
plt.ylabel("Age" + u"\u2192")

plt.axhline(mean_age, color='r', label='mean_age' + " : " + str(round(mean_age, 2)))

plt.bar(range(len(age)), np.sort(age), label='age', color='g')

plt.legend()

plt.show()
```

```python
# let's club same age together and form counts of the age

from collections import Counter

age_counts = Counter(age)

xs = range(max(age))
ys = [age_counts[x] for x in xs]
```

```python
plt.bar(xs, ys, label='Age of Players')

plt.axvline(mean_age, color='r', label='mean_age' + " : " + str(round(mean_age, 2)))

plt.title("Age Distribution Of Fifa Players")
plt.xlabel("Age Of Player" + u"\u2192")
plt.ylabel("Count Of Players" + u"\u2192")

plt.legend()

plt.show()
```

# Median

```python
def median(data):
    if len(data) % 2 == 0:
        return sorted(data)[(len(data) + 1) // 2 ]
    else:
        return sorted(data)[len(data) // 2]
```

```python
median_age = median(age)
```

```python
plt.bar(xs, ys, label='Age of Players')

plt.axvline(mean_age, color='y', label='mean_age' + " : " + str(round(mean_age, 2)))

plt.axvline(median_age, color='g', label='medain_age' + " : " + str(median_age))

plt.title("Age Distribution Of Fifa Players")
plt.xlabel("Age Of Player" + u"\u2192")
plt.ylabel("Count Of Players" + u"\u2192")

plt.legend()

plt.show()
```

```python
dispersion = max(age) - min(age)

print(dispersion)
```

```python
def variance(data):
    n = len(data)
    deviations = [x - mean_age for x in data]
    dev_square = [x**2 for x in deviations]
    return sum(dev_square) / (n - 1)
```

```python
var = variance(age)
print(var)
```

```python
import math

def std(data):
    return math.sqrt(variance(data))
```

```python
std(age)
```
