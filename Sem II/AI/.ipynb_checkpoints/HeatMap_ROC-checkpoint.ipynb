```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, shuffle=True)
```

```python
from sklearn import decomposition
nof_pca = 2

pca = decomposition.PCA(n_components=nof_pca, svd_solver='full').fit(X_train)

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
nohn = 10
clf = MLPClassifier(hidden_layer_sizes=(nohn,), solver='lbfgs', activation='tanh',
                    batch_size = 15, verbose = False, early_stopping = True).fit(X_train_pca, y_train)
```

```python
y_pred = clf.predict(X_test_pca)
```

```python
# comparing the results in tabular fromat

import pandas as pd

# setting pandas options to display more data
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 10)

data = {'target' : y_test}

worksheet = pd.DataFrame(data)

worksheet['mlp_y_pred'] = y_pred

print(worksheet)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

# Accuracy measures for MLP

```python
# cross validation
import numpy as np
from sklearn.model_selection import cross_val_score

# just for demostration purpose
# For accurate prediction, make sure PCA of input data are passed
# as the classifier was trained with PCA data
X_pca = pca.transform(X)
cross_val_result = cross_val_score(clf, X_pca, y, cv=4)

print("MLP cross validation scores : ", cross_val_result)
print("MLP cross validation average accuracy : ", np.mean(cross_val_result))
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
# confusion matrix as a heatmap using seaborn

import seaborn as sns
import matplotlib.pyplot as plt

classes = np.unique(y_test)
xticklabels = classes.astype(int).astype(str).tolist()
yticklabels = classes.astype(int).astype(str).tolist()

ax = plt.subplot()
# ax.set_xticks(classes)
# ax.set_yticks(classes)

# annot = True to annotate cells
# ftm='g' to disable scientific notation
colormap = sns.color_palette("Reds")
sns.heatmap(cm, annot=True, fmt='g', ax=ax, xticklabels= xticklabels, yticklabels=yticklabels, cmap=colormap)

# labels
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True Labels')
ax.set_title('Confusion Matrix')
```

```python
# Preaparing data for ROC curve display

y_score = clf.fit(X_train_pca, y_train).predict_proba(X_test_pca)

from sklearn.preprocessing import LabelBinarizer

label_binarizer = LabelBinarizer().fit(y_train)
y_onehot_test = label_binarizer.transform(y_test)
print('Test Shape', y_onehot_test.shape) # (n_samples, n_classes)

class_of_interest = 2.0
class_id = np.flatnonzero(label_binarizer.classes_ == class_of_interest)

print('Corresponding Class Id', class_id)
```

```python
# ROC cure for oe class

from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_predictions(y_onehot_test[:, clas I am doing in the middle andss_id],
                                y_score[:, class_id],
                                name=f"{class_of_interest} vs the rest",
                                color='Darkorange',)

plt.plot([0,1], [0,1], "k--", label="chance level (AUC=0.5)")
plt.axis("square")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"One-vs-Rest ROC curves:\nClass {class_of_interest} vs. rest of the class")
plt.legend()
plt.show()
```

```python
# ROC AUC score

from sklearn.metrics import roc_auc_score

roc_auc_score(y_onehot_test[:,class_id],
             y_score[:, class_id])
```

# Adam Boosting

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

# Addabooster classifier
adbost = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1),
                            n_estimators=10, learning_rate=0.1).fit(X_train, y_train)
```

```python
# Use the model/predictor to predict
Y_pred = adbost.predict(X_test)
```

```python
X_pca = pca.transform(X)
cross_val_result = cross_val_score(adbost, X_pca, y, cv=4)

print("MLP cross validation scores : ", cross_val_result)
print("MLP cross validation average accuracy : ", np.mean(cross_val_result))
```
